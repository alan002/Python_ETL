2 Padronizar a coluna 'itens': Limpar e padronizar os dados na coluna 'itens'.
Arredondar 'valor total': Arredondar os valores da coluna 'valor_total' para duas casas decimais.

3 Classificar fraude: Criar uma nova coluna para categorizar o nível de fraude.

4 Criar timestamp da compra: Converter a data da compra para timestamp.

5 Extrair ano e mês da compra: Criar uma coluna com o ano e o mês da compra.

6 Categorizar formas de pagamento: Criar uma coluna com categorias padronizadas para as formas de pagamento.

7 Finalizar a tarefa: Apresentar os resultados das transformações realizadas no dataset.




```python
import pandas as pd
import unidecode

# Carregar dataset
arquivo = 'https://drive.google.com/uc?id=1mjfSme6yG9111vSy9zUIsO5CBLM0ndXw'
df = pd.read_csv(arquivo)

# 1. Padronização da coluna 'itens'
df['itens'] = df['itens'].str.lower()
df['itens'] = df['itens'].apply(lambda x: unidecode.unidecode(str(x)))
df['itens'] = df['itens'].str.replace(r'[^\w\s]', '', regex=True)
df
```

```python
# 2. Arredondar 'valor_total' para 2 casas decimais sem afetar nulos
df['valor_total'] = df['valor_total'].apply(lambda x: round(x, 2) if pd.notnull(x) else x)
```

```python
# 3. Criar coluna 'fraude_suspeita'
def classificar_fraude(valor):
    if pd.isnull(valor):
        return 'indefinido'
    elif valor < 0.3:
        return 'baixa'
    elif valor < 0.7:
        return 'moderado'
    else:
        return 'alta'

df['fraude_suspeita_float'] = pd.to_numeric(df['fraude_suspeita'], errors='coerce')
df['fraude_suspeita2'] = df['fraude_suspeita_float'].apply(classificar_fraude)
```

```python
# 4. Criar coluna 'pagamento_ts'
df['data_compra_ts'] = pd.to_datetime(df['data_compra'], errors='coerce').apply(lambda x: x.timestamp() if pd.notnull(x) else None)
```

```python
# 5. Criar coluna 'ano_mes_compra'
df['ano_mes_compra'] = pd.to_datetime(df['data_compra'], errors='coerce').dt.to_period('M').astype(str)
df
```

```python
# 6. Criar coluna 'forma_pagamento_cat'
mapeamento = {
    'pix': 'pix',
    'boleto': 'boleto',
    'cartão de crédito': 'cartao_credito',
    'paypal': 'paypal'
}

df['forma_pagamento_cat'] = df['meio_pagamento'].apply(lambda x: mapeamento.get(str(x).lower(), 'outro'))

```
