(Hands on) Carregue o dataset de pagamentos (https://drive.google.com/uc?id=1Cmi7ChwivaFO5C9qNM1G8oQ-hk1RORgt) e faça
as seguintes operações: * Transforme as colunas 'created_at' e 'confirmed_at' para timestamp.
* A partir da data, crie novas colunas para representar o ano, mês, dia e hora do pagamento.
* Crie uma nova coluna com a data no formato 'dd/mm/yyyy' e outra no formato 'Ano-mês'
* Calcule a diferença entre 'created_at' e 'confirmed_at' e classifique a transação de acordo com os dados a seguir:
"Instantânea" se a diferença for menor que 5 minutos; "Normal" se for entre 5 e 30 minutos; "Demorada" se for acima
de 30 minutos. (dica: use dt.total_seconds() depois de subtrair as datas) **negrito**

```python
import pandas as pd
import datetime

pd.Timestamp.now()
arquivo = 'https://drive.google.com/uc?id=1Ooex2Jei9sEcuuzhwlpJUKLKX4zQD-Kr'
df = pd.read_csv(arquivo)

# Função para classificar a transação com base na diferença de tempo
def classify_transaction(time_diff):
    if pd.isna(time_diff):
        return 'Pending' # Ou qualquer outro rótulo apropriado para valores ausentes
    seconds = time_diff.total_seconds()
    if seconds < 300:  # 5 minutes = 300 seconds
        return 'Instantânea'
    elif 300 <= seconds <= 1800:  # 30 minutes = 1800 seconds
        return 'Normal'
    else:
        return 'Demorada'

# Garantir que 'created_at' e 'confirmed_at' sejam objetos de data e hora
df['created_at'] = pd.to_datetime(df['created_at'])
df['confirmed_at'] = pd.to_datetime(df['confirmed_at'])

# Calcular a diferença entre 'confirmed_at' e 'created_at'
df['time_difference'] = df['confirmed_at'] - df['created_at']

# Classify the transaction based on the time difference
df['transaction_classification'] = df['time_difference'].apply(classify_transaction)

# Transforme as colunas 'created_at' e 'confirmed_at' para timestamp.
# Converter colunas de data e hora em timestamp (em segundos desde a época) # Use errors='coerce' para manipular valores NaT potenciais em 'confirmed_at'
df['created_at_timestamp'] = df['created_at'].apply(lambda x: x.timestamp() if pd.notna(x) else None)
df['confirmed_at_timestamp'] = df['confirmed_at'].apply(lambda x: x.timestamp() if pd.notna(x) else None)


# Exibir as colunas relevantes para mostrar o resultado
display(df[['created_at', 'created_at_timestamp', 'confirmed_at', 'confirmed_at_timestamp', 'time_difference', 'transaction_classification']].head())

#A partir da data, crie novas colunas para representar o ano, mês, dia e hora do pagamento.

df['ano_pagamento'] = df['created_at'].dt.year
df['mes_pagamento'] = df['created_at'].dt.month
df['dia_pagamento'] = df['created_at'].dt.day
df['hora_pagamento'] = df['created_at'].dt.hour

display(df[['created_at', 'ano_pagamento', 'mes_pagamento', 'dia_pagamento', 'hora_pagamento']].head())


#* Crie uma nova coluna com a data no formato 'dd/mm/yyyy' e outra no formato 'Ano-mês'
df['dat_formatada_dd_mm_yyyy'] = df['created_at'].dt.strftime('%d/%m/%Y')
df['dat_formatada_ano_mes'] = df['created_at'].dt.strftime('%Y-%m')

display(df[['created_at', 'dat_formatada_dd_mm_yyyy', 'dat_formatada_ano_mes']].head())
```

Carregue o arquivo dataset_bootcamp.csv através do link:

Execute as seguintes operações:
1. Padronização de escrita da coluna itens: Remova todos os acentos gráficos, cedilhas e caracteres
especiais da coluna itens. Tudo deve ficar em minúsculas, sem ~, ç, á, etc.
2. Arredonde todos os valores da coluna valor_total para duas casas decimais (float). Cuidado para
não afetar os valores nulos ou inválidos.
3. Crie uma coluna chamada risco_fraude seguindo as seguintes regras
a. Alto se fraude_suspeita >= 0.7
b. Moderado se fraude_suspeita entre 0.3 e 0.7
c. Baixo se fraude_suspeita < 0.3
d. Indefinido se fraude_suspeita estiver vazio
4. Crie uma nova coluna chamada data_compra_ts, transformando data_compra em timestamp
inteiro
5. Crie uma nova coluna ano_mes_compra no formato YYYY-MM.
6. A coluna meio_pagamento pode conter variações como pix, PIX, Pix, etc. Padronize todos os
valores para:
a. 'pix', 'boleto', 'cartao_credito', 'paypal'
b. valores fora disso devem ser convertidos para "outro"

```python
import pandas as pd
arquivo = 'https://drive.google.com/uc?id=1mjfSme6yG9111vSy9zUIsO5CBLM0ndXw'
df4 = pd.read_csv(arquivo)
#Padronização de escrita da coluna itens: Remova todos os acentos gráficos, cedilhas e caracteres
#especiais da coluna itens. Tudo deve ficar em minúsculas, sem ~, ç, á, etc.
df4['itens'] = df4['itens'].str.normalize('NFKD').str.encode('ascii', errors='ignore').str.decode('utf-8').str.lower()

#Arredonde todos os valores da coluna valor_total para duas casas decimais (float). Cuidado para não afetar os valores nulos ou inválidos.
df4['valor_total'] = df4['valor_total'].round(2)

#Crie uma coluna chamada risco_fraude seguindo as seguintes regras a. Alto se fraude_suspeita >= 0.7 b. Moderado se fraude_suspeita entre 0.3 e 0.7 c. Baixo se fraude_suspeita < 0.3 d. Indefinido se fraude_suspeita estiver vazio
def riscoFraude(fraude_suspeita):
    if pd.isna(fraude_suspeita):
        return "Indefinido"
    elif fraude_suspeita < 0.3:
        return "Baixo"
    elif fraude_suspeita >=0.3 and fraude_suspeita < 0.7:
        return "Moderado"
    elif fraude_suspeita >= 0.7:
        return "Alto"

# outra forma de fazer o lambda
# dados['risco_fraude'] = dados['fraude_suspeita'].apply(lambda x: "Indefinido" if pd.isna(x) else ("Baixo" if x < 0.3 else ("Moderado" if 0.3 <= x < 0.7 else "Alto")))
# df['valor_total'] = df['valor_total'].applay(lambda x: round(x,2) if pd.notnull(x) else x)

df4["risco_fraude"]  = df4["fraude_suspeita"].apply(lambda x: riscoFraude(x))
dadosPrint = df4[["fraude_suspeita","risco_fraude"]]
display(dadosPrint.head())

#Crie uma nova coluna chamada data_compra_ts, transformando data_compra em timestamp inteiro
df4['data_compra_ts'] = pd.to_datetime(df4['data_compra']).astype(int)

#Crie uma nova coluna ano_mes_compra no formato YYYY-MM.
df4['ano_mes_compra'] = pd.to_datetime(df4['data_compra_ts'], unit='ns').dt.strftime('%Y-%m')

#A coluna meio_pagamento pode conter variações como pix, PIX, Pix, etc. Padronize todos os valores para: a. 'pix', 'boleto', 'cartao_credito', 'paypal' b. valores fora disso devem ser convertidos para "outro"
df4['meio_pagamento'] = df4['meio_pagamento'].str.lower().replace({
    'pix': 'pix',
    'boleto': 'boleto',
    'cartao de credito': 'cartao_credito',
    'paypal': 'paypal'
}).fillna('outro') # Substituir valores que não estão no mapa por 'outro'

df4['meio_pagamento'].head()
```

```python
#Função lambda para classificar risco de fraude
lambda_riscoFraude = lambda fraude_suspeita: "Indefinido" if pd.isna(fraude_suspeita) else (
    "Baixo" if fraude_suspeita < 0.3 else (
        "Moderado" if 0.3 <= fraude_suspeita < 0.7 else "Alto"
    )
)

# Aplique a função lambda para criar a coluna 'risco_fraude_lambda'
df4['risco_fraude_lambda'] = df4['fraude_suspeita'].apply(lambda_riscoFraude)

# Exibir as colunas relevantes para comparar os resultados
display(df4[['fraude_suspeita', 'risco_fraude', 'risco_fraude_lambda']].head())
```
