## **Criando data frames**

### Utilizando a URL para pegar todas as informações dos países no continente europeu (https://restcountries.com/v3.1/region/europe) crie um data frame que contenha o nome comum e oficial em inglês e na língua local, a subregião, a moeda, a língua e a capital de cada um deles.

```python
#Passo 1: importando a biblioteca requests, pandas e json
import requests as r
import pandas as pd
import json

#Passo 2: Montando as variáveis para a URL completa
base_url = 'https://restcountries.com/v3.1' #endereço raíz da api
endpoint = '/region/' #rota específica da API (endpoint)
region = 'europe'

#Passo 3: construindo a URL completa da requisição
url = base_url + endpoint + region
print("URL Final:",url) #verificando qual foi a url final
```

```python
#Passo 4: Criando requisição e adicionando a uma variável
resposta = r.get(url) #criando requisição
print('Status code:', resposta.status_code) #verificando status code
```

```python
#Passo 5: verificando e conhecendo o output
data = resposta.json() #salvando o output da requisição em uma variável
print("Verificando header", resposta.headers, '\n') # metadados da requisição, pode conter informações relevantes como o tipo de arquivo que recebemos
print(resposta.text, '\n') #Exibe a resposta bruta da API como string (texto puro)
print('chaves do arquivo', data[0].keys(), '\n') #Visualiza as chaves (campos) do primeiro registro do JSON
print(data[0], '\n') #Mostra o conteúdo completo do primeiro item da resposta
print(json.dumps(resposta.json(), indent=2, ensure_ascii=False)) #ver o conteúdo formatado da resposta
for chave, valor in data[0].items(): #Percorre todas as chaves e valores do primeiro país retornado
    print(chave, "--", valor)
```

```python
#Passo 6: trazer as informações que queremos para formato de tabela/data frame
#Passo 6.1: Fazer uma lista para armazenar os dados tratados
paises = []
#Passo 6.2: iterar sobre todos os países retornados
for pais in data:
  nome_comum = pais.get('name', {}).get('common', 'n/a'),#pega o nome comum em inglês
  nome_local = pais.get('name', {}).get('official', 'n/a')#pega nome oficial em inglês
  nome_nativo_dict = pais.get('name', {}).get('nativeName', 'n/a')#tenta pegar nome oficial na língua local
#  print('nome nativo dict:', nome_nativo_dict, '\n')
  if nome_nativo_dict:
    first_lang = list(nome_nativo_dict.values())[0]
#    print('first lang:', first_lang, '\n')
    nome_ntv = first_lang.get('official', 'n/a')
#    print('nome ntv:', nome_ntv, '\n')
  else:
    nome_ntv = 'n/a'
  capital = pais.get('capital', ['n/a'])[0]
  subregiao = pais.get('subregion', 'n/a')
  moeda = pais.get('currencies', {})
  linguas = pais.get('languages', {})
  paises.append({
      'Nome Comum (Inglês)': nome_comum,
      'Nome Oficial (Inglês)': nome_local,
      'Nome Oficial (Língua Local)': nome_ntv,
      'Capital': capital,
      'Subregião': subregiao,
      'Moeda': moeda,
      'Linguas': linguas
  })

df = pd.DataFrame(paises)
df
```

### - Utilizando a URL para pegar as informações de Brasil, Argentina, África do Sul e Japão (https://restcountries.com/v3.1/alpha?codes=bra,zaf,arg,jpn)crie um data frame somente para as informações do Japão. Crie um data frame diferente juntando a informação de todos eles. Para ambos os data frames precisamos do nome oficial do país em inglês, o tld, a abreviação da moeda, seu símbolo e nome, região, subregião, línguas e população.

```python
#Passo 1: importando a biblioteca REQUESTS (comunicação com páginas e aplicações web)
import requests as r
import pandas as pd
import json

#Passo 2: Montando as variáveis para a URL completa
base_url = 'https://restcountries.com/v3.1' #endereço raíz da api
endpoint = '/alpha' #rota específica da API (endpoint)
params = '?codes=bra,zaf,arg,jpn' #parâmetros de filtro: campos que queremos receber na resposta (filtro do JSON)

#Passo 3: construindo a URL completa da requisição
url = base_url + endpoint + params
print("URL Final:",url) #verificando qual foi a url final
```

```python
#Passo 4: fazendo a requisição
resposta = r.get(url) #criando variável com requisição da url
print('Status code:', resposta.status_code) #testando se requisição teve sucesso
```

```python
#Passo 5: verificando e conhecendo o output
data = resposta.json() #salvando o output da requisição em uma variável
print("Verificando header", resposta.headers, '\n') # metadados da requisição, pode conter informações relevantes como o tipo de arquivo que recebemos
print(resposta.text, '\n') #Exibe a resposta bruta da API como string (texto puro)
print('chaves do arquivo', data[0].keys(), '\n') #Visualiza as chaves (campos) do primeiro registro do JSON
print(data[0], '\n') #Mostra o conteúdo completo do primeiro item da resposta
print(json.dumps(resposta.json(), indent=2, ensure_ascii=False)) #ver o conteúdo formatado da resposta
for chave, valor in data[0].items(): #Percorre todas as chaves e valores do primeiro país retornado
    print(chave, "--", valor)
```

```python
#Passo 6: Montando o data frame
#Passo 6.1: Listas para armazenar os dados
dados_todos = [] #dados de todos os países
dados_japao = [] #dados somente do japão

# Passo 6.2: Iterar sobre os países retornados
for pais in data:
    nome_oficial = pais.get("name", {}).get("official", "N/A") #Nome oficial
    tld = pais.get("tld", ["N/A"])[0] #TLD (domínio de topo, ex: .br)
    moedas = pais.get("currencies", {}) #Moeda (abreviação, nome e símbolo)
    if moedas:
        abrev = list(moedas.keys())[0]
        moeda_info = moedas[abrev]
        nome_moeda = moeda_info.get("name", "N/A")
        simbolo_moeda = moeda_info.get("symbol", "N/A")
    else:
        abrev = nome_moeda = simbolo_moeda = "N/A"

    regiao = pais.get("region", "N/A") #continente
    subregiao = pais.get("subregion", "N/A") #subregião
    linguas = pais.get("languages", {}) #línguas
    populacao = pais.get("population", 0) #População

    # Construção do dicionário com os dados
    registro = {
        "Nome Oficial": nome_oficial,
        "TLD": tld,
        "Moeda (abrev)": abrev,
        "Moeda (nome)": nome_moeda,
        "Moeda (símbolo)": simbolo_moeda,
        "Região": regiao,
        "Sub-região": subregiao,
        "Línguas": linguas,
        "População": populacao
    }

    # Verifica se é o Japão
    if pais.get("cca3") == "JPN":
        dados_japao.append(registro)

    # Adiciona ao DataFrame geral
    dados_todos.append(registro)

# Passo 5: Criar os DataFrames
df_japao = pd.DataFrame(dados_japao)
df_todos = pd.DataFrame(dados_todos)

# Exibir
print(df_japao)
print(df_todos)
```
